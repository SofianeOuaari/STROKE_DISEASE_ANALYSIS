```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np 
import cufflinks as cf
import plotly.offline as py  
import plotly.graph_objs as go
cf.go_offline
py.init_notebook_mode
df_train=pd.read_csv("train_2v.csv",sep=",") 
df_test=pd.read_csv("test_2v.csv",sep=",")
df_train.head(15)
```

```python
df_clean_train=df_train.dropna()
print(len(df_clean_train))
print(len(df_train))
```

```python
g=df_clean_train["gender"].unique().tolist() 
print(g)
ev=df_clean_train["ever_married"].unique().tolist()
print(ev)
wt=df_clean_train["work_type"].unique().tolist()
print(wt)
rt=df_clean_train["Residence_type"].unique().tolist()
print(rt)
smt=df_clean_train["smoking_status"].unique().tolist() 
print(smt)
```

```python
df_clean_train["gender"]=df_clean_train["gender"].astype("category").cat.codes
df_clean_train["ever_married"]=df_clean_train["gender"].astype("category").cat.codes 
df_clean_train["work_type"]=df_clean_train["work_type"].astype("category").cat.codes
df_clean_train["Residence_type"]=df_clean_train["Residence_type"].astype("category").cat.codes 
df_clean_train["smoking_status"]=df_clean_train["smoking_status"].astype("category").cat.codes
```

```python
df_clean_train.head(15)
```

```python
df_clean_train.columns
features=['gender', 'age', 'hypertension', 'heart_disease', 'ever_married',
       'work_type', 'Residence_type', 'avg_glucose_level', 'bmi',
       'smoking_status']
x=df_clean_train[features]
y=df_clean_train["stroke"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.tree import DecisionTreeClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,stratify=y,random_state=315)
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
print(accuracy_score(y_test,y_pred))
```

```python
df_test.isnull().any()
df_cleaning_test=df_test.dropna()
df_cleaning_test["ever_married"]=df_cleaning_test["ever_married"].astype("category").cat.codes
df_cleaning_test["gender"]=df_cleaning_test["gender"].astype("category").cat.codes
df_cleaning_test["work_type"]=df_cleaning_test["work_type"].astype("category").cat.codes
df_cleaning_test["Residence_type"]=df_cleaning_test["Residence_type"].astype("category").cat.codes
df_cleaning_test["smoking_status"]=df_cleaning_test["smoking_status"].astype("category").cat.codes
```
